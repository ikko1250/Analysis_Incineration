```python
import pandas as pd

df = pd.read_csv("/home/ubuntu/cur/program/Analyisis_incineration/EU/ロンバルディア_プロット用_modified_1.0.csv")
df.head()
```

```python
numeric_cols = df.select_dtypes(include="number").columns.tolist()
print(f"Numeric columns ({len(numeric_cols)}): {numeric_cols}")
print("Missing values per numeric column:")
print(df[numeric_cols].isna().sum())
df[numeric_cols].describe().T
```

```python
target_col = "R1係数"
corr_with_target = df[numeric_cols].corr()[target_col].dropna().sort_values(key=lambda s: s.abs(), ascending=False)
corr_with_target.to_frame(name="corr_with_target")
```

```python
import numpy as np
from IPython.display import display
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import LeaveOneOut, cross_val_score

correlation_threshold = 0.9
max_features = 3

# Pick up to three weakly correlated predictors to avoid singular matrices.
selected_features = []
for feature, corr_value in corr_with_target.items():
    if feature == target_col:
        continue
    if len(selected_features) >= max_features:
        break
    if all(abs(df[feature].corr(df[keep])) <= correlation_threshold for keep in selected_features):
        selected_features.append(feature)

model_data = df[selected_features + [target_col]].dropna()
X = model_data[selected_features]
y = model_data[target_col]

model = LinearRegression()
model.fit(X, y)

train_r2 = model.score(X, y)

loo = LeaveOneOut()
mae_scores = cross_val_score(model, X, y, cv=loo, scoring="neg_mean_absolute_error")
mse_scores = cross_val_score(model, X, y, cv=loo, scoring="neg_mean_squared_error")

metrics = pd.DataFrame({
    "value": [train_r2, -mae_scores.mean(), np.sqrt(-mse_scores).mean()]
}, index=["train_r2", "loocv_mae", "loocv_rmse"])

coef_df = pd.DataFrame({
    "coefficient": model.coef_
}, index=selected_features)

print(f"Selected features: {selected_features}")
print(f"Intercept: {model.intercept_:.4f}")
display(metrics)
display(coef_df)
```
