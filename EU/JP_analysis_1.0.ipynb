```python
def calculate_lcv_from_composition(V, W):
    """
    三成分から低位発熱量を独自計算
    
    Parameters:
    V: 可燃分（質量比、0〜1）
    W: 水分（質量比、0〜1）
    
    Returns:
    HI: 低位発熱量 [kJ/kg]
    """
    # HI (kcal/kg) = 4,500V - 600W
    hi_kcal = 4500 * V - 600 * W
    
    # HI(kJ/kg) = HI(kcal/kg) × 4.1868
    hi_kj = hi_kcal * 4.1868
    
    return hi_kj

# 「三成分水分％」「三成分可燃分％」を0〜1に変換
V = df_waste["三成分可燃分％"] / 100
W = df_waste["三成分水分％"] / 100

# 計算して新しいカラムに代入
df_waste["低位発熱量(独自)kJ/kg"] = calculate_lcv_from_composition(V, W)

# 結果の先頭5行を表示
df_waste[["三成分可燃分％", "三成分水分％", "低位発熱量(独自)kJ/kg"]].head()
```

```python
def iqr_filter(df, col, k=1.5):
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - k * iqr
    upper = q3 + k * iqr
    mask = (df[col] >= lower) & (df[col] <= upper)
    # 範囲・件数を表示
    print(f"外れ値判定範囲: {lower:.2f} ～ {upper:.2f}")
    print(f"外れ値として置き換え対象となる件数: {(~mask).sum()} 件")
    return mask, lower, upper

columns_to_filter = [
    "低位発熱量(実測値)kJ/kg",
    "低位発熱量(独自)kJ/kg", 
    "低位発熱量(計算値)kJ/kg"  
]

for col in columns_to_filter:
    # 0をnanに置き換え
    df_waste.loc[df_waste[col] == 0, col] = np.nan
    # IQRで外れ値除去
    mask, lower, upper = iqr_filter(df_waste, col, k=1.5)
    df_waste.loc[~mask, col] = np.nan
    print(f"{col} の外れ値除去後の値の範囲: {df_waste[col].min():.2f} ～ {df_waste[col].max():.2f}")

sns.displot(df_waste, x="低位発熱量(実測値)kJ/kg", kde=True)
sns.displot(df_waste, x="低位発熱量(独自)kJ/kg", kde=True)
sns.displot(df_waste, x="低位発熱量(計算値)kJ/kg", kde=True)
```

```python
# 最終的に使用する低位発熱量を決定するカラムと、その出所を示すカラムを作成します。
# 優先順位: 実測値 > 計算値 > 独自

# 条件リスト
conditions = [
    df_waste["低位発熱量(実測値)kJ/kg"].notna(),
    df_waste["低位発熱量(計算値)kJ/kg"].notna(),
    df_waste["低位発熱量(独自)kJ/kg"].notna()
]

# 値の選択リスト
value_choices = [
    df_waste["低位発熱量(実測値)kJ/kg"],
    df_waste["低位発熱量(計算値)kJ/kg"],
    df_waste["低位発熱量(独自)kJ/kg"]
]

# 出所を示す文字列の選択リスト
source_choices = [
    "実測値",
    "計算値",
    "独自"
]

# np.selectを使用して新しいカラムを作成
df_waste["最終低位発熱量kJ/kg"] = np.select(conditions, value_choices, default=np.nan)
df_waste["採用熱量"] = np.select(conditions, source_choices, default="該当なし")

# 「該当なし」の行を実測値の中央値で補完
# 中央値を計算
median_lcv_actual = df_waste["低位発熱量(実測値)kJ/kg"].median()

# 該当なしのマスクを作成
no_value_mask = df_waste["採用熱量"] == "該当なし"

# 値と採用熱量カラムを更新
df_waste.loc[no_value_mask, "最終低位発熱量kJ/kg"] = median_lcv_actual
df_waste.loc[no_value_mask, "採用熱量"] = "実測値中央値"


# 結果の確認
print("採用熱量の内訳:")
print(df_waste["採用熱量"].value_counts())

# 関連するカラムの先頭を表示して確認
display(df_waste[[
    "低位発熱量(実測値)kJ/kg",
    "低位発熱量(計算値)kJ/kg",
    "低位発熱量(独自)kJ/kg",
    "最終低位発熱量kJ/kg",
    "採用熱量"
]].head())

# 最終的な低位発熱量の分布を可視化
sns.displot(df_waste, x="最終低位発熱量kJ/kg", hue="採用熱量", kind="kde", fill=True)
plt.title("最終低位発熱量の分布（採用熱量別）")
plt.show()
```

```python
def calculate_energy_efficiency(df_calc):
    """
    エネルギー効率（R1）を計算する関数
    """
    # 1. Ep (年間エネルギー生産量) の計算
    # 電気エネルギー: 発電量 [MWh] → [GJ] (1 MWh = 3.6 GJ, 係数2.6をかける)
    electricity_gj = df_calc['発電能力総発電量（実績値）MWh'].fillna(0) * 3.6 * 2.6
    
    # 熱エネルギー: 余熱利用量 [MJ] → [GJ]
    # 外部熱供給分は商用熱として係数1.1をかける
    external_heat_gj = df_calc['余熱利用量（実績値）うち外部熱供給量MJ'].fillna(0) / 1000 * 1.1
    # 場内利用分は係数1.0として扱う（商用ではないため）
    total_heat_mj = df_calc['余熱利用量（実績値）余熱利用量MJ'].fillna(0)
    external_heat_mj = df_calc['余熱利用量（実績値）うち外部熱供給量MJ'].fillna(0)
    internal_heat_gj = (total_heat_mj - external_heat_mj) / 1000 * 1.0
    
    df_calc['Ep_electricity_GJ'] = electricity_gj
    df_calc['Ep_external_heat_GJ'] = external_heat_gj
    df_calc['Ep_internal_heat_GJ'] = internal_heat_gj
    df_calc['Ep_total_GJ'] = electricity_gj + external_heat_gj + internal_heat_gj
    
    # 2. Ew (廃棄物の持つエネルギー) の計算
    # 年間処理量 [t] × 低位発熱量 [kJ/kg] → [GJ]
    lcv = df_calc['最終低位発熱量kJ/kg']
    annual_waste_kg = df_calc['年間処理量t/年度'] * 1000  # t → kg
    df_calc['Ew_GJ'] = annual_waste_kg * lcv / 1000000  # kJ → GJ
    
    # 3. Ef と Ei の推定 (追加燃料と外部エネルギー輸入)
    # データがないため0と仮定
    df_calc['Ef_GJ'] = 0
    df_calc['Ei_GJ'] = 0
    
    # 4. エネルギー効率の計算 (R1 formula)
    # 分子: Ep - (Ef + Ei)
    numerator = df_calc['Ep_total_GJ'] - (df_calc['Ef_GJ'] + df_calc['Ei_GJ'])
    # 分母: 0.97 * (Ew + Ef)
    denominator = 0.97 * (df_calc['Ew_GJ'] + df_calc['Ef_GJ'])
    
    # ゼロ除算を避ける
    df_calc['energy_efficiency'] = np.where(
        denominator > 0,
        numerator / denominator,
        np.nan
    )
    return df_calc

# 関数を適用してエネルギー効率を計算
df_waste = calculate_energy_efficiency(df_waste.copy())

# 結果の要約統計量を表示
df_waste[['energy_efficiency']].describe()
```
