```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df_waste = pd.read_csv("/home/ubuntu/cur/program/Analyisis_incineration/EU/1焼却施設_2022csv.csv")
df_waste.head()
```

```python
for c in df_waste.columns:
    print(c)
```

```python
# '施設の種類' カラムのデータでヒストグラムを作成します。
# 日本語フォントの文字化けを防ぐために japanize_matplotlib をインポートします。
import japanize_matplotlib

# '施設の種類' カラムの値の数をカウント
facility_counts = df_waste['施設の種類'].value_counts()

# 棒グラフで可視化
plt.figure(figsize=(12, 8))
sns.barplot(y=facility_counts.index, x=facility_counts.values, orient='h', palette='viridis')
plt.title('施設の種類ごとの数', fontsize=16)
plt.xlabel('施設数', fontsize=12)
plt.ylabel('施設の種類', fontsize=12)
plt.tight_layout() # レイアウトを調整
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import japanize_matplotlib

# '使用開始年度'を数値に変換し、無効な値はNaNにする
df_waste['year'] = pd.to_numeric(df_waste['使用開始年度'], errors='coerce')

# yearがNaNでないデータのみを対象とする
df_plot = df_waste.dropna(subset=['year'])

# 年度を整数に変換
df_plot['year'] = df_plot['year'].astype(int)

# 年度と施設の種類でクロス集計
cross_tab = pd.crosstab(df_plot['year'], df_plot['施設の種類'])

# 積み上げ棒グラフの作成
cross_tab.plot(kind='bar', stacked=True, figsize=(15, 8), colormap='tab10')

plt.title('施設の種類ごとの年間建設数', fontsize=16)
plt.xlabel('使用開始年度', fontsize=12)
plt.ylabel('施設数', fontsize=12)
plt.legend(title='施設の種類')
plt.tight_layout()
plt.show()
```
